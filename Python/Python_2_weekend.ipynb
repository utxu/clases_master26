```python
def process_a(lines):
    result = []
    
    for line in lines:
        if len(line) > 1:
          result.append(last_word(line))
    return result

process_a(lines)
```

```python
def process_b(lines):
    result = []
    
    for line in lines:
        
        if len(line) > 1:
            
          result.append(line[0])
        
    return result

process_b(lines)
```

```python
def process(f, lines):
    result = []
    
    for line in lines:
        
        if len(line) > 1:
            
          result.append(f(line))
        
    return result

process(len, lines)
process(last_word, lines)
process(str.upper, lines)

```

```python
def busca_nombre(lista, letra):
    
    res_lita = []
    
    for nombre in lista:
        
        if letra.lower() in nombre.lower(): #lo hago case insensitive
            
            res_lista.append(nombre)

    return res_lista        
```
