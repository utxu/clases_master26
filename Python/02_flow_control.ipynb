```python
sequence =[1, 2, None,4, None, 6]
```

```python
total=0
for value in sequence:
    if value is None:
        continue
    total+=value
    
total
```

```python
total=0
for value in sequence:
    if value is None:
        break
    total+=value
    
total
```
